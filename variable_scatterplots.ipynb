```python
# import libraries possible needed
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
```

```python
# import dataset
spotify = pd.read_csv('Spotify Music Data.csv')
spotify.head()
```

```python
ols_object = smf.ols(formula = 'pop~bpm+nrgy+dnce+dB+live+val+dur+acous+spch', data = spotify)
mlr_spotify = ols_object.fit()
```

### bpm vs pop

```python
ax = spotify.plot(x = 'bpm', y = 'pop', kind = 'scatter',figsize=(10, 6),xlabel = 'bpm', 
              ylabel = 'pop')
```
